```python
import pyttsx3 as pp
import wikipedia 
import os
import speech_recognition as sr 
import datetime 
import webbrowser
import smtplib 

engine = pp.init()
voices = engine.getProperty('voices')

engine.setProperty('voice', voices[0].id)
def speak(input):
    engine.say(input)
    engine.runAndWait()

# speak('divyanshu') #testing
def wishme():
    hour = int(datetime.datetime.now().hour)
    if hour >=0 and hour <12:
        speak("good morning")
    elif hour>=12 and hour <18:
        speak("Good afternoon")
    else:
        speak("Good evening")
    speak("i am your chatbot, what is your good name")


# wishme()
 
def tc():
    r = sr.Recognizer()
    with sr.Microphone() as source:
        print('listening...')
        r.adjust_for_ambient_noise(source)
        r.pause_threshold = 1
        audio = r.listen(source)
    try: 
        print('Recognizing..')
        resp = r.recognize_google(audio)
        print(resp)
    except Exception as e:
        print('say that again please')
        return 'None'
    return resp

def sendEmail(to,content):
    server = smtplib.SMTP('smtp.gmail.com', 587)
    server.ehlo()
    server.starttls()
    server.login(' ','######')
    server.sendmail(' ', to, content)
    server.close()

# tc() #testing

wishme()
# for name 
name = tc()
speak(f'hello {name}, please tell me how can i help you sir')


while True:
    resp = tc().lower()
    # to search anything on wikipedia
    if 'Wikipedia' in resp:
        speak("searching wikipedia..")
        resp = resp.replace("Wikipedia", "")
        results = wikipedia.summary(resp, sentences =2)
        speak('According to wikipedia')
        print(results)
        speak(results)
    # opening some web applications
    elif 'google' in resp:
        webbrowser.open('google.com')
    elif 'open facebook' in resp:
        webbrowser.open('facebook.com')
    elif 'open youtube' in resp:
        webbrowser.open('youtube.com')
    elif 'open whatsapp' in resp:
        webbrowser.open('whatsapp.com')
    elif 'time' in resp:
        strtime = datetime.datetime.now().strftime('%H:%M:%S')
        speak(f'Sir, time is :{strtime}')
    # to open an app in the pc
    elif 'open zoom' in resp:
        OpenZoom = 'C:\\Users\\divya\\AppData\\Roaming\\Zoom\\bin\\Zoom.exe'  #target of zoom in properties
        os.startfile(OpenZoom)

    elif 'email' in resp:
        try:
            content = tc()
            to = "divyanshuagrawal162@gmail.com"
            sendEmail(to, content)
            speak("Email has been sent")
        except Exception as e:
            speak('sorry, i am not able to send this email')
    else:
        # say anything or exit, quit such keywords to stop chatbot
        last_note = 'thank you, and feel free to ask anything'
        speak(last_note)   
        print(last_note) 
        break
```
